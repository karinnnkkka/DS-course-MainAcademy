<b><div style="text-align: center;font-size: 24px;">Конкурсний відбір</div></b><br>

Напишіть програму, яка виводить список хорошистів та відмінників у класі.

**Формат вхідних даних**
На вхід програмі подається натуральне число n, далі слідує n рядків із прізвищем школяра та її оцінкою кожної з них.

**Формат вихідних даних**
Програма має вивести спочатку всі введені рядки з прізвищами та оцінками учнів у тому самому порядку. Потім слідує порожній рядок, а потім виводяться рядки з прізвищами та оцінками хорошистів та відмінників (у тому ж порядку).

**Примітка 1.** Оцінка учня – це натуральне число від 1 до 5.

**Примітка 2.** Гарантується, що у класі є хоча б один хороший – володар оцінки 4, або відмінник - отримав 5.

<b>Sample Input</b><br>
5  
Круглов 4   
Кузнецов 5  
Федин 4  
Тарасов 2  
Словецкий 3  

<b>Sample Output</b><br> 
Круглов 4  
Кузнецов 5  
Федин 4  
Тарасов 2  
Словецкий 3  

Круглов 4  
Кузнецов 5  
Федин 4  





```python
n = int(input('Введіть кількість учнів: '))
names = []
marks =[]

for i in range(n):
    i = input('Введіть прізвище учня: ')
    names.append(i)
    
for j in range(n):
    j = int(input('Введіть оцінку учня : '))
    marks.append(j)
    
journal = dict(zip(names, marks))

for key, value in journal.items():
    print(key, value)

print('\n')

for key, value in journal.items():
    if value >= 4:
        print(key, value)
```

Напишіть програму, яка зчитує натуральне число n і виводить перші n чисел послідовності Трибоначчі.

**Формат вхідних даних**
На вхід до програми подається одне число (n≤100) – кількість членів послідовності.

**Примітка.** Послідовність Трибоначчі – послідовність натуральних чисел, де кожне наступне число є сумою трьох попередніх:




<b>Sample Input</b><br>
10  

<b>Sample Output</b><br> 
1 1 1 3 5 9 17 31 57 105



```python
n = int(input())
a, b, c = (1, 1, 1)

print(a)
print(b)
print(c)
for _ in range(3,n,1):
    print(a + b + c)
    
    c = a + b + c
    b = c - a - b
    a = c - b - a
```
