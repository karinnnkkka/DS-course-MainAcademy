- 0 на zero;

**Примітка 2.** Ваша програма повинна ігнорувати будь-які символи, які не вказані в наведеній вище таблиці.

На вхід до програми подається рядок тексту. Напишіть програму, яка виводить слово, яке зустрічається найрідше, без урахування регістру. Якщо таких слів кілька, виведіть те, що є меншим у лексикографічному порядку.

**Примітка 1.** Програма не повинна бути чутливою до регістру, слова apple та Apple повинна сприймати як однакові.

**Примітка 2.** Слово – послідовність літер. Крім слів у тексті можуть бути пробіли та розділові знаки .,!?:;-, які потрібно ігнорувати. Інших символів у тексті немає.

<b>Sample Input 1</b><br>
home sweet home

<b>Sample Output 1</b><br> 
sweet

<br><br>
<b>Sample Input 2</b><br>
My dear, here we must run as fast as we can, just to stay in place. And if you wish to go anywhere you must run twice as fast as that.

<b>Sample Output 2</b><br> 
and


```python
line = input('Enter the sentence: ').upper().strip(',.?!-:;').split(' ')
dictionary = {}

for i in line:
    dictionary[i] = line.count(i)
    
mini = min(dictionary.values())
res = {}
for key, value in dictionary.items():
    if value == mini:
        res[key] = value
        
print(min(res))
```

<b><div style="text-align: center;font-size: 24px;">Анаграми</div></b><br>

Анаграма – слово (словосполучення), утворене шляхом перестановки букв, що становлять інше слово (або словосполучення). Наприклад, англійські слова evil та live – це анаграми.

На вхід до програми подаються дві пропозиції. Напишіть програму, яка визначає, чи є вони анаграмами чи ні. Ваша програма повинна ігнорувати регістр символів, розділові знаки та пробіли.

<b>Sample Input 1</b><br>
Вижу зверей  
Живу резвей

<b>Sample Output 1</b><br> 
YES

<br><br>
<b>Sample Input 2</b><br>
Когда увидимся  
тогда и свидимся

<b>Sample Output 2</b><br> 
NO








```python
linef = input('Enter the first sentence: ').upper().strip(',.?!-:; ')
lines = input('Enter the second sentence: ').upper().strip(',.?!-:; ')

dictf = dict(zip(sorted(linef), range(len(linef))))
dicts = dict(zip(sorted(lines), range(len(lines))))

if dictf == dicts:
    print('YES')
else:
    print('NO')
```

**Формaт вхідних даних**
Програма отримує на вхід кількість країн n. Далі йде n рядків, кожен рядок починається з назви країни, потім йдуть назви міст цієї країни. У наступному рядку записано число , далі йдуть m запитів - назви якихось міст, з перерахованих вище.

**Формат вихідних даних**
Програма має вивести назву країни, в якій знаходиться це місто відповідно до прикладу.

<b>Sample Input</b><br>
2

```python
dictionary = {}

n = int(input('Enter the number of countries: '))

for i in range(n):
    country, *city = input('Enter county and cities: ').split(' ')
    dictionary[country] = city
    
m = int(input('Enter the number of cities: '))

for j in range(m):
    j = input('Enter the name of city: ')
    for country, city in dictionary.items():
        if j in city:
            print(country)
```

<b><div style="text-align: center;font-size: 24px;">Секретне слово</div></b><br>

Напишіть програму для розшифровування секретного слова шляхом частотного аналізу.

**Формат вхідних даних**
У першому рядку задано зашифроване слово. У другому рядку задано одне ціле число n – кількість літер у словнику. У наступних  рядках записано, скільки разів конкретна буква алфавіту зустрічається у цьому слові – <літера>: <частота>.

**Формат вихідних даних**
Програма має вивести дешифроване слово.

**Примітка.** Гарантується, що частоти букв не повторюються.

<b>Sample Input</b><br>
p!p!p?  
3         
а: 3  
н: 2  
с: 1  

<b>Sample Output</b><br> 
ананас










```python
cipher = input('Enter a sentence to decipher: ')
n = int(input('Enter the amount of letters in dictionary: '))
dictionary = {}

for i in range(n):
    value, key = input("Enter a letter and frequency separated by ':' : ").split(': ')
    dictionary[int(key)] = value

for _ in cipher:
    print(dictionary[cipher.count(_)], end ='')
```

<b><div style="text-align: center;font-size: 24px;">Небезпечний вірус 😈</div></b><br>

У файлову систему комп'ютера, проник небезпечний вірус і зламав контроль прав доступу до файлів. Кажуть, вірус написав один із студентів курсу Python для початківців.

Для кожного файлу відомо, з якими діями можна звертатися до нього:

- запис W (write, доступ на запис у файл);
- читання R (read, доступ читання з файла);
- запуск X (execute, запуск виконання файла).

Напишіть програму для відновлення прав доступу до файлів. Ваша програма для кожного запиту повинна буде повертати OK, якщо виконується допустима операція, і Access denied, якщо операція неприпустима.

**Формат вхідних даних**
Програма отримує кількість файлів на вхід n, що містяться у файловій системі комп'ютера. Далі йде n рядків, на кожній ім'я файлу та допустимі з ним операції, розділені символом пробілу. У наступному рядку записано число m — кількість запитів до файлів, а потім m рядків із запитами виду операція файл. Одному тому файлу може бути адресовано будь-яку кількість запитів.

**Формат вихідних даних**
Програма має вивести для кожного з m запитів в окремому рядку Access denied або OK.

<b>Sample Input 1</b><br>
5        
my_pycode.exe W X  
log_n X W R  
ave R  
lucky_m W R  
dnsss.py W  
6  
execute ave  
read dnsss.py  
write log_n       
execute log_n  
read ave  
write my_pycode.exe  

<b>Sample Output 1</b><br> 
Access denied  
Access denied  
OK  
OK   
OK  
OK  

<br><br>
<b>Sample Input 2</b><br>
2  
marvel_movies X  
dc_com X R  
2  
execute dc_com  
write dc_com  

<b>Sample Output 2</b><br> 
OK  
Access denied  











```python
n = int(input('Enter the number of files: '))

operations = {'R': 'read', 'W': 'write', 'X': 'execute'}
dictionary = {}

for i in range(n):
    i = input('Enter file and commands: ').split(' ')
    dictionary[i[0]] = [operations[_] for _ in i[1:]]
    
m = int(input('Enter the number of tries: '))

for j in range(m):
    j = input('Enter the request: ').split(' ')
    if j[0] in dictionary[j[1]]:
        print('OK')
    else:
        print('Access denied')
```
