<b><div style="text-align: center;font-size: 24px;">Слід матриці</div></b><br>

Після квадратної матриці називається сума елементів головної діагоналі. Напишіть програму, яка виводить слід заданої квадратної матриці.

<b>Sample Input</b><br>
3  
1 2 3  
4 5 6  
7 8 9  

<b>Sample Output</b><br> 
15






```python
n = int(input('Enter the dimension of the matrix: '))
a = [[int(input('Enter the element: ')) for i in range(n)] for j in range(n)]

d = 0

for i in range(n):
    for j in range(n):
        if(i == j):
            d += a[i][j]

print(d)
```

<b><div style="text-align: center;font-size: 24px;">Більше середнього</div></b><br>

Напишіть програму, яка виводить кількість елементів квадратної матриці в кожному рядку, більших за середній арифметичний елементів даного рядка.

<b>Sample Input</b><br>
4  
1 2 3 4  
5 6 3 15   
0 2 3 1  
5 2 8 5   

<b>Sample Output</b><br> 
2  
1  
2  
1  









```python
n = int(input('Enter the dimension of the matrix: '))
a = [[int(input('Enter the element: ')) for i in range(n)] for j in range(n)]

for i in range(n): 
    counter = 0
    avg = 0
    for k in range(n):
        temp = a[i][k] / n
        avg += temp
    for j in range(n):
        if a[i][j] > avg:
            counter += 1
    print('\n', counter)
```

<b><div style="text-align: center;font-size: 24px;">Максимальний в області 1</div></b><br>

Напишіть програму, яка виводить максимальний елемент у заштрихованій області квадратної матриці.



<b>Sample Input</b><br>
3
1 4 5
6 7 8
1 1 6  

<b>Sample Output</b><br> 
7




```python
n = int(input('Enter the dimension of the matrix: '))
a = [[int(input('Enter the element: ')) for i in range(n)] for j in range(n)]

res =[]

for i in range(n):
    for j in range(n):
        if i >= j :
            res.append(a[i][j])

print(max(res))
```

<b><div style="text-align: center;font-size: 24px;">Максимальний в області 2 🌶️</div></b><br>

Напишіть програму, яка виводить максимальний елемент у заштрихованій області квадратної матриці.



<b>Sample Input</b><br>
3  
1 4 5  
6 7 8  
1 1 6    

<b>Sample Output</b><br> 
8




```python
n = int(input('Enter the dimension of the matrix: '))
a = [[int(input('Enter the element: ')) for i in range(n)] for j in range(n)]

res = []

for i in range(n):
    for j in range(n):
        if ((i >= j) & (i <= n-j-1)) | ((i <= j) & (i >= n-j-1)) | (i==j) | (i + j +1 == n):
            res.append(a[i][j])

print(max(res))
```

<b><div style="text-align: center;font-size: 24px;">Максимум у таблиці</div></b><br>

На вхід програмі подаються два натуральні числа n та m - кількість рядків і стовпців у матриці, потім n рядків по m цілих чисел у кожному, відокремлених символом пробілу.

Напишіть програму, яка знаходить індекси (рядок та стовпець) першого входження максимального елемента.

<b>Sample Input</b><br>
3  
4  
0 3 2 4  
2 3 5 5  
5 1 2 3   

<b>Sample Output</b><br> 
1 2





```python
n = int(input('Enter the number of rows: '))
m = int(input('Enter the number of columns: '))
a = [[int(input('Enter the element: ')) for i in range(m)] for j in range(n)]

maxel = a[0][0]
x = 0
y = 0

for i in range(n):
    for j in range(m):
        if(a[i][j] > maxel):
            maxel = a[i][j]
            x = i
            y = j
            
print(x,y)
```

<b><div style="text-align: center;font-size: 24px;">Обмін стовпців</div></b><br>

Напишіть програму, яка змінює місцями стовпці у матриці.

**Формат вхідних даних**
На вхід програмі на різних рядках подаються два натуральні числа n та m — кількість рядків і стовпців у матриці, потім елементи матриці рядкове через пробіл, потім числа i та j – номери стовпців, які підлягають обміну.

**Формат вихідних даних**
Програма повинна вивести вказану таблицю із заміненими стовпцями.

<b>Sample Input</b><br>
3  
5  
12 14 11 13 24  
22 24 21 23 14  
32 34 31 33 34  
0 2  

<b>Sample Output</b><br> 
11 14 12 13 24  
21 24 22 23 14  
31 34 32 33 34  




```python
n = int(input('Enter the number of rows: '))
m = int(input('Enter the number of columns: '))

a = [[int(input('Enter the element: ')) for i in range(m)] for j in range(n)]

i, j = int(input('Enter first column to swap: ')), int(input('Enter the second column to swap: '))

for k in range(n):
    a[k][i], a[k][j] = a[k][j], a[k][i]
    
for k in range(n):
    for p in range(m):
        print(a[k][p], end =' ')
    print('\n')
```

<b><div style="text-align: center;font-size: 24px;">Обмін діагоналей</div></b><br>

Дано квадратну матрицю чисел. Напишіть програму, яка змінює місцями елементи, що стоять на головній та побічній діагоналі, при цьому кожен елемент повинен залишитися в тому ж стовпці (тобто у кожному стовпці потрібно поміняти місцями елемент на головній діагоналі та побічній діагоналі).

<b>Sample Input</b><br>
3  
1 2 3  
4 5 6  
7 8 9   

<b>Sample Output</b><br> 
7 2 9   
4 5 6  
1 8 3  




```python
n = int(input('Enter dimension of matrix: '))

a = [[float(input('Enter the element: ')) for i in range(n)] for j in range(n)]

for i in range(n):
    a[i][i], a[n-i-1][i] = a[n-i-1][i], a[i][i]
            
            
for k in range(n):
    for p in range(n):
        print(a[k][p], end =' ')
    print('\n')
```

<b><div style="text-align: center;font-size: 24px;">Ходи коня</div></b><br>

На шахівниці 8×8 стоїть кінь. Напишіть програму, яка відзначає положення коня на дошці та всі клітини, які б'є кінь. Клітку, де стоїть кінь, позначте англійською буквою N, клітини, які б'є кінь, відзначте символами *, решту клітин заповніть крапками.

**Формат вхідних даних**
На вхід програмі подаються координати коня на шахівниці в шахової нотації (тобто у вигляді e4, де спочатку записується номер стовпця (літера від a до h, зліва направо), потім номер рядка (цифра від 1 до 8, знизу догори)).

**Формат вихідних даних**
Програма повинна вивести на екран зображення дошки, розділяючи елементи пробілами.

**Примітка.** Шахова дошка має вигляд:




<b>Sample Input</b><br>
b6

<b>Sample Output</b><br> 




```python
x, y = input(('Enter a letter of point: ')), int(input('Enter a number of point: '))
n = 8
board = [['.'] * n for _ in range(n)]
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
y = n - y
x = letters.index(x)
board[y][x] = 'N'

for i in range(n):
    for j in range(n):
        if abs(i-y)*abs(j-x) == 2:
            board[i][j] = '*'
            
for i in range(n):
    print(*board[i])
```
