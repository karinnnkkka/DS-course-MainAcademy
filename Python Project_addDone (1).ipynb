<b><div style="text-align: center;font-size: 24px;">Шифр Цезаря</div></b><br>

Шифр Цезаря (шифр зсуву) — одне із найпростіших і найвідоміших методів шифрування. Шифр Цезаря - це вид [шифру підстановки](https://ru.wikipedia.org/wiki/%D0%A8%D0%B8%D1%84%D1%80_%D0%BF%D0%BE%D0%B4%D1%81%D1%82%D0%B0%D0%BD%D0%BE%D0%B2%D0%BA%D0%B8), в якому кожен символ у відкритому тексті замінюється символом, що знаходиться на деякому постійному числі позицій ліворуч або правіше за нього в алфавіті.

>Шифр підстановки — метод шифрування із заміною елементів вихідного відкритого тексту іншими відповідно до деякого правила.

Наприклад, у шифрі зі зрушенням праворуч 3 позиції символ A замінюється символом D, символ B символом E, і так далі, до символу Z, що замінюється символом C.

>Шифр названий на честь римського полководця [Гая Юлія Цезаря](https://ru.wikipedia.org/wiki/%D0%93%D0%B0%D0%B9_%D0%AE%D0%BB%D0%B8%D0%B9_%D0%A6%D0%B5%D0%B7%D0%B0%D1%80%D1%8C), який використовував його для таємного листування зі своїми генералами.

Шифр Цезаря легко зламується і майже ніякого практичного застосування.

Шифр Цезаря може бути легко зламаний навіть у випадку, коли хакер знає лише зашифрований текст. Можна розглянути дві ситуації:

1. Зломщик знає (або припускає), що використовувався простий шифр підстановки, але не знає, що це схема Цезаря.
2. Зломщик знає, що використовувався шифр Цезаря, але знає значення зсуву.

У першому випадку шифр може бути зламаний із застосуванням методу [частотного аналізу](https://ru.wikipedia.org/wiki/%D0%A7%D0%B0%D1%81%D1%82%D0%BE%D1%82%D0%BD%D1%8B%D0%B9_%D0%B0%D0%BD%D0%B0%D0%BB%D0%B8%D0%B7). Використовуючи цей метод, зломщик, ймовірно, швидко помітить регулярність у вирішенні і зрозуміє, що шифр, що використовується, - шифр Цезаря.

У другому випадку злом шифру ще простіший. Існує не так багато варіантів значень зсуву (26 для англійської мови, 32 для російської), всі вони можуть бути перевірені перебором.

<b><div style="text-align: center;font-size: 24px;">Примітки</div></b><br>
**Примітка 1.** Перетворення, що використовується в шифрі Цезаря, зазвичай позначають як ROT N, де N — зсув, ROT — скорочення від слова ROTATE, в даному випадку «циклічний зсув». Наприклад, позначення ROT 2 позначає зсув на 2 позиції, тобто, "а" перетворюється на "в", "б" на "г", і так далі, і в кінці "ю" перетворюється на "а" а "я" - на "б".

**Примітка 2.** Число різних змін залежить від довжини алфавіту:

- для російської мови можливо 32 різних перетворення (перетворення ROT 0 та ROT 33 зберігають вихідний текст, а далі починаються вже повторення); 
- для англійської мови можливі 25 різних перетворень (перетворення ROT 0 та ROT 26 зберігають вихідний текст, а далі починаються вже повторення).

**Примітка 3.** Неалфавітні символи — розділові знаки, пробіли, цифри — не змінюються.

**Примітка 4.** Природним розвитком шифру Цезаря став [шифр Віженера](https://ru.wikipedia.org/wiki/%D0%A8%D0%B8%D1%84%D1%80_%D0%92%D0%B8%D0%B6%D0%B5%D0%BD%D0%B5%D1%80%D0%B0).

<b><div style="text-align: center;font-size: 24px;">Аве, Цезар 🌶️</div></b><br>
На вхід до програми подається рядок тексту англійською мовою, в якому потрібно зашифрувати всі слова. Кожне слово рядка слід зашифрувати за допомогою шифру Цезаря (циклічного зсуву на довжину цього слова). Рядкові літери при цьому залишаються малими, а великі - великими. Гарантується, що між різними словами є одна прогалина.

**Формат вхідних даних**  
На вхід до програми подається рядок тексту англійською мовою.

**Формат вихідних даних**  
Програма має вивести зашифрований текст відповідно до умови завдання.

**Примітка.** Символи, які не є англійськими літерами, не змінюються.

**Sample Input 1:**  
Day, mice. "Year" is a mistake!

**Sample Output 1:**  
Gdb, qmgi. "Ciev" ku b tpzahrl!

<br>

**Sample Input 2:**  
my name is Python!

**Sample Output 2:**  
oa reqi ku Veznut!

```python
alphabetl = 'abcdefghijklmnopqrstuvwxyz'
alphabetup = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

line = input('Enter your line to cipher: ')
shift = line.split(' ')

length = []

for i in shift:
    for j in i:
        if j.isalpha() == 0:
            i = i.replace(j, '')
    length.append(i)


j = 0


for i in line:
    n = len(length[j])
    if i in alphabetl:
        index = alphabetl.index(i)
        if index + n < len(alphabetl):
            print(alphabetl[index + n], end = '')
        else:
            print(alphabetl[index + n - len(alphabetl)], end='')
    elif i in alphabetup:
        index = alphabetup.index(i)
        if index + n < len(alphabetup):
            print(alphabetup[index + n], end = '')
        else:
            print(alphabetup[index + n - len(alphabetup)], end='')
    else:
        print(i, end ='')
        if i == ' ':
            j += 1
```
