<b><div style="text-align: center;font-size: 24px;">–°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Ç—Ä—å–æ—Ö üå∂Ô∏è</div></b><br>
    
–ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –≤–ø–æ—Ä—è–¥–∫–æ–≤—É—î —Ç—Ä–∏ —á–∏—Å–ª–∞ –≤—ñ–¥ –±—ñ–ª—å—à–æ—ó –¥–æ –º–µ–Ω—à–æ—ó.

<b>Sample Input</b><br>

129  
135  

<b>Sample Output</b><br> 
135  
132  
129  



```python
a = float(input("Enter your number: "))
b = float(input("Enter your number: "))
c = float(input("Enter your number: "))

temp = 0
while ( (a > b) | (b > c) | (a > c)):
    if (a > b):
        temp = a
        a = b
        b = temp
    elif (b > c):
        temp = b
        b = c
        c = temp
    else:
        temp = a
        a = c
        c = temp

lst = [a, b , c]
for i in lst:
    print(i)
```

<b><div style="text-align: center;font-size: 24px;">–Ñ–≤–∫–ª—ñ–¥–æ–≤–∞ –≤—ñ–¥—Å—Ç–∞–Ω—å</div></b><br>
    
–ù–∞ –ø–ª–æ—â–∏–Ω—ñ –µ–≤–∫–ª—ñ–¥–æ–≤–∞ –≤—ñ–¥—Å—Ç–∞–Ω—å –º—ñ–∂ –¥–≤–æ–º–∞ —Ç–æ—á–∫–∞–º–∏ ($x_{1}$, $y_{1}$) —ñ ($x_{2}$, $y_{2}$) –≤–∏–∑–Ω–∞—á–∞—î—Ç—å—Å—è —Ç–∞–∫ $\sqrt{(x_{1}- x_{2})^{2} + (y_{1}-y_{2})^{2}}$

<b>Sample Input</b><br>
5.5  
2.4  
4.9  
6.3  

<b>Sample Output</b><br> 
3.9458839313897713 





```python
x1 = float(input("Enter your x1: "))
x2 = float(input("Enter your x2: "))
y1 = float(input("Enter your y1: "))
y2 = float(input("Enter your y2: "))

evc = ((x1 - x2)**2 +(y1 - y2)**2)**0.5
print("\n", evc)
```

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞ 2.</b> –ú–æ–¥—É–ª—å math –º—ñ—Å—Ç–∏—Ç—å –≤–±—É–¥–æ–≤–∞–Ω—É —Ñ—É–Ω–∫—Ü—ñ—é radians(), —è–∫–∞ –ø–µ—Ä–µ–≤–æ–¥–∏—Ç—å –∫—É—Ç —ñ–∑ –≥—Ä–∞–¥—É—Å—ñ–≤ —É –∫—É—Ç —É —Ä–∞–¥—ñ–∞–Ω–∞—Ö.
