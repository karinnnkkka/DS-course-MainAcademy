<b><div style="text-align: center;font-size: 24px;">Зворотне число</div></b><br>
    
Напишіть програму, яка зчитує з клавіатури одне число і виводить йому зворотне. Якщо при цьому введене з клавіатури число - нуль, вивести "Зворотного числа не існує" (без лапок).




<b>Sample Input 1</b><br>
2.5

<b>Sample Output 1</b><br> 
0.4

<br><br>
<b>Sample Input 2</b><br>
0

<b>Sample Output 2</b><br> 
Зворотного числа не існує




```python
a = float(input("Введіть число: "))

if (a == 0):
    print("\nЗворотнього числа не існує")
else:
    print("\n", 1/a)
```

<b><div style="text-align: center;font-size: 24px;">Сортування трьох 🌶️</div></b><br>
    
Напишіть програму, яка впорядковує три числа від більшої до меншої.

<b>Sample Input</b><br>
132  
129  
135  

<b>Sample Output</b><br> 
135  
132  
129  



```python
a = float(input("Enter your number: "))
b = float(input("Enter your number: "))
c = float(input("Enter your number: "))

temp = 0
while ( (a > b) | (b > c) | (a > c)):
    if (a > b):
        temp = a
        a = b
        b = temp
    elif (b > c):
        temp = b
        b = c
        c = temp
    else:
        temp = a
        a = c
        c = temp

lst = [a, b , c]
for i in lst:
    print(i)
```

<b><div style="text-align: center;font-size: 24px;">Абсолютна сума</div></b><br>

Дані п'ять чисел $a_{1}$, $a_{2}$, $a_{3}$, $a_{4}$, $a_{5}$. Напишіть програму, яка обчислює суму їх модулів $\left | a_{1} \right |$ + $\left | a_{2} \right |$ + $\left | a_{3} \right |$ + $\left | a_{4} \right |$ + $\left | a_{5} \right |$.

<b>Sample Input</b><br>
5.4  
33  
-1232  
-3.889  
6   

<b>Sample Output</b><br> 
1280.289 


```python
suma = 0

for i in range (5):
    i = float(input("Enter your number: "))
    if( i < 0):
        i *= (-1)
    suma += i
    
print("\n", suma)
```

<b><div style="text-align: center;font-size: 24px;">Євклідова відстань</div></b><br>
    
На площині евклідова відстань між двома точками ($x_{1}$, $y_{1}$) і ($x_{2}$, $y_{2}$) визначається так $\sqrt{(x_{1}- x_{2})^{2} + (y_{1}-y_{2})^{2}}$

<b>Sample Input</b><br>
5.5  
2.4  
4.9  
6.3  

<b>Sample Output</b><br> 
3.9458839313897713 





```python
x1 = float(input("Enter your x1: "))
x2 = float(input("Enter your x2: "))
y1 = float(input("Enter your y1: "))
y2 = float(input("Enter your y2: "))

evc = ((x1 - x2)**2 +(y1 - y2)**2)**0.5
print("\n", evc)
```

<b><div style="text-align: center;font-size: 24px;">Тригонометричний вираз</div></b><br>
    
Напишіть програму, яка обчислює значення тригонометричного виразу за заданою кількістю градусів x.








<b>Примітка 1.</b> Тригонометричні функції набувають аргументу в радіанах. Щоб перевести градуси в радіани, скористайтесь формулою



<b>Примітка 2.</b> Модуль math містить вбудовану функцію radians(), яка переводить кут із градусів у кут у радіанах.

<b>Sample Input</b><br>
30.0 

<b>Sample Output</b><br> 
1.6993587371177719 



```python
import math as mt

x = float(input("Enter x in degrees: "))
x = mt.radians(x)

print("\n", mt.sin(x) + mt.cos(x) + mt.tan(x)**2)
```
