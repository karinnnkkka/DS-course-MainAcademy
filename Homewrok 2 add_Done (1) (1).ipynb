<b><div style="text-align: center;font-size: 24px;">Церемонія зважування</div></b><br>
    
Відома вага боксера-аматора (ціле число). Відомо, що вага така, що боксер може бути віднесений до однієї з трьох вагових категорій:

- Легка вага – до 60 кг;
- Перша напівсередня вага – до 64 кг;
- Напівсередня вага – до 69 кг.
Напишіть програму, яка визначає, в якій категорії виступатиме цей боксер.

<b>Sample Input</b><br>
55  

<b>Sample Output</b><br> 
Легка вага

```python
weight = int(input("Введіть вагу боксера: "))

if (weight <= 0):
    print("Хибне значення ваги")
elif (weight < 60):
    print("Легка вага")
elif (60 <= weight < 64):
    print("Перша напівсередня вага")
elif ( 64 <= weight <69):
    print("Напівсередня вага")
else:
    print("Боксер не може бути віднесений до жодної з категорій")
```

Напишіть програму, яка приймає на вхід число і в залежності від умов виводить текст YES або NO.

Умови:
- якщо число непарне, вивести «YES»;
- якщо число парне в діапазоні від 2 до 5 (включно), вивести «NO»;
- якщо число парне в діапазоні від 6 до 20 (включно), вивести «YES»;
- якщо число парне і більше 20, то вивести NO

<b>Sample Input</b><br>
1

<b>Sample Output</b><br> 
YES

```python
a = int(input("Enter your number: "))
if (a % 2 == 0):
    if(2 <= a <= 5):
        print("\nNO")
    elif ( 6 <= a <= 20):
        print("\nYES")
    elif (a > 20):
        print("\nNO")
else:
    print("\nYES")
```
