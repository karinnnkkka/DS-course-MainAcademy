### Завдання 1

Для трьох введених натуральних чисел (користувач вводить числа через пробіл) знайдіть мінімальне та максимальне

```python
numbers = input('Enter 3 numbers separated by space: ')
numbers = numbers.split(' ')
mini = numbers[0]
maxi = numbers[1]
for i in numbers:
    if int(i) < int(mini):
        mini = i
    elif int(i) > int(maxi):
        maxi = i
    else: 
        continue

print('Максимальне: ', maxi)
print('Мінімальне: ', mini)
```

### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруємо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,10) for i in range(1000)]


def minmaxa(lists):
    mina = lists[0]
    maxa = lists[1]
    
    for _ in lists:
        if(_ < mina):
            mina = _
        elif(_ > maxa):
            maxa = _
    print('Мінімальне:', mina)
    print('Максимальне:', maxa)

def middle(lists):
    avg = 0
    for _ in lists:
        avg += _
    avg = avg/len(lists)
    #print('Середнє значення послідовності: ', avg)
    
    return avg
    
def mediana(lists):
    med = 0
    lists.sort()
    if len(lists)%2 == 0:
        med += ((lists[len(lists)//2]+lists[len(lists)//2 - 1])/2)
    else:
        med += (lists[len(lists)//2])
        
    print('Медіана: ', med)

def frequency(lists):
    quantity = {i: lists.count(i) for i in lists}
    
    print('Кількість повторень: ', quantity)
        
        
def standard_deviation(lists):
    mean = middle(lists)
    res = 0
    for i in lists:
        a = (i - mean) ** 2
        res += a
        
    res = (res/(len(lists)-1))**0.5
    
    print('Середнє квадратичне відхилення: ', res)
    
print(list_with_random_numbers)

minmaxa(list_with_random_numbers)
print('Середнє значення послідовності: ', middle(list_with_random_numbers))
mediana(list_with_random_numbers)
frequency(list_with_random_numbers)
standard_deviation(list_with_random_numbers)
```
